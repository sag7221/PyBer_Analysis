```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
# Read CSV
csv_path = os.path.join("Resources","got.csv")
got_df = pd.read_csv(csv_path)
got_df.head()
got_df.isna().sum()
```

```python
# Get attacker and defender king data
# Clean up Data
#Find all Nan
got_attacker_series = got_df['attacker_king'].value_counts()
got_attacker_series
```

```python
# Get total battle data
got_defender_series = got_df['defender_king'].value_counts()
got_defender_series
```

```python
#Combine the series
total_battles_series = got_attacker_series.add(got_defender_series, fill_value=0)
total_battles_series
```

```python
# Configure plot and ticks
total_battles_series.plot(kind="bar")


# Set textual properties


# Show plot


# Resize plot to display labels
```
